```python
import pandas as pd
import numpy as np
import math
from sklearn.linear_model import LinearRegression
```

```python
df = pd.DataFrame(pd.read_excel("data_5.xlsx"))
```

```python
an1=[]
diffv=[]
diffx=[]
vn1=[]
for ind in df.index:
  if ind%2==0:
    a = 1
  else:
    an1.append(df['Acceleration'][ind])
    vn1.append(df['Speed'][ind])
    diffx.append(df['Position'][ind-1]-df['Position'][ind])
    diffv.append(df['Speed'][ind-1]-df['Speed'][ind])
```
